```python
import numpy as numpy
import time

def procees_image(images: np.ndarray) -> np.ndarray:
    time.sleep(1)
    return 255 - images

start_time = time.time()
images = [np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8) for _ in range(10)]
processed_images = procees_image(np.array(images))
print("Processed images shape:", processed_images.shape)
end_tine = time.time()

print("Processing time:", end_tine - start_time)
```
